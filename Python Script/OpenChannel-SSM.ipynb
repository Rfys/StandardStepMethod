```python
'''
---
All codes are available in StandardStepMethod Module
To use it, please copy the file into your project folder and import it
---
Importing Matplotlib to configure the result to your desire is also suggested
---
'''

import StandardStepMethod as ssm
import matplotlib.pyplot as plt
```

```python
'''
There are 2 functions wrapper that you can use
First one is Calc_NormalCriticalDepth that return Normal Depth and Critical Depth
of a channel with known discharge and channel characteristic
Below is the example of the wrapper usage
'''


Yc, Yn = ssm.Calc_NormalCriticalDepth(
    discharge = 6, 
    base_width = 1, 
    left_slope = 1, 
    right_slope = 1, 
    manning = 0.025,
    slope = 0.02
)
```

```python
'''
The second on is UniformChannel
This wrapper gives you the water profile alongside with its normal depth and critical depth with same channel characteristic throughout the stream
The result is formed into Pandas DataFrame that you can custumize further with reference to 
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html.
Here an example of its usage
'''

profile1 = ssm.uniformChannel(
    Discharge = 6, 
    Slope = 0.002, 
    Base = 1, 
    Manning = 0.025, 
    M1 = 1, 
    M2 = 1, 
    Boundary_Depth = 1.1, 
    Initial_x = 0, 
    delta_x = 1, 
    Distance = 250, 
    BaseDepthAtBoundary = 0, 
    Direction = 1
)

profile1
```

```python
'''
There's also additional plt.plot wrapper to build the graph quickly
Here's the example:
'''


ssm.createPlot(profile1, fig_width=15, fig_heigth=5)
```

```python
profile2 = ssm.uniformChannel(
    Discharge = 6, 
    Slope = 0.08, 
    Base = 1, 
    Manning = 0.025, 
    M1 = 1, 
    M2 = 1, 
    Boundary_Depth = 1.065, 
    Initial_x = 0, 
    delta_x = 0.1, 
    Distance = 200, 
    BaseDepthAtBoundary = 0, 
    Direction = -1
)

profile2
```

```python
ssm.createPlot(profile2)
plt.xlim(0,15)
plt.ylim(-2, 2)
```

```python
'''
The createPlot wrapper can also accept more than one analysis in an array like structure such as below
'''

ssm.createPlot([profile1, profile2])
plt.xlim(-20, 20)
plt.ylim(-1, 2)
plt.savefig("img.jpg")
```

```python
'''
There is also a conjugate and alternate depth version of ploting if you wish
However, these functions only accept one analyze result at a time
'''

ssm.createPlot(profile1)
ssm.alternatePlot(profile1)
ssm.conjugatePlot(profile1)
plt.legend()
plt.show()
```
