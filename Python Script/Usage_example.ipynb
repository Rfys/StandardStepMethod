```python
import StandardStepMethod as ssm
import matplotlib.pyplot as plt

Hulu = ssm.uniformChannel(
    Discharge = 6, 
    Slope = 0.001, 
    Base = 1, 
    Manning = 0.025, 
    M1 = 1, 
    M2 = 1, 
    Boundary_Depth = 4, 
    Initial_x = 0, 
    delta_x = 10, 
    Distance = 2000, 
    BaseDepthAtBoundary = 0, 
    Direction = 1
)

Hulu.set_index('X Position')
```

```python
Hilir = ssm.uniformChannel(
    Discharge = 6, 
    Slope = 0.001, 
    Base = 1, 
    Manning = 0.025, 
    M1 = 1, 
    M2 = 1, 
    Boundary_Depth = 2.5, 
    Initial_x = 1500, 
    delta_x = 5, 
    Distance = 1500, 
    BaseDepthAtBoundary = 1500*(-0.001), 
    Direction = 1
)

Hilir.set_index('X Position')
```

```python
Pintu = ssm.uniformChannel(
    Discharge = 6, 
    Slope = 0.001, 
    Base = 1, 
    Manning = 0.025, 
    M1 = 1, 
    M2 = 1, 
    Boundary_Depth = 0.4859, 
    Initial_x = 0, 
    delta_x = 0.05, 
    Distance = 40,
    BaseDepthAtBoundary = 0, 
    Direction = -1
)

Pintu.set_index("X Position")
```

```python
LoncatAir = ssm.checkHydraulicJump(
    Upstream    = Pintu,
    Downstream  = Hilir
)

LoncatAir
```

```python
profile = [Hulu, Hilir, Pintu]
ssm.createPlot(profile)
ssm.alternatePlot(Hulu)
ssm.conjugatePlot(Pintu)
plt.legend()
plt.axvline(0, ymin=0.36, c='r')
plt.xlim(-100, 100)
plt.savefig("Pintu Air.jpg")
```

```python
profile = [Hulu, Hilir, Pintu]
ssm.createPlot(profile)
plt.axvline(0, ymin=0.37, c='r')
plt.savefig("Saluran.jpg")
```

```python
ssm.createPlot([Hulu, LoncatAir])
ssm.conjugatePlot(Pintu)
plt.legend()
plt.axvline(0, ymin=0.37, c='r')
plt.xlim(-100, 100)
plt.show()
```
